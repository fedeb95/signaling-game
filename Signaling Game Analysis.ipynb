```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('signaling_2_4.csv')
data.tail()
```

Plot success rate at the end of each try.

```python
data[data['Run'] == 1000].set_index('Try')['SuccessRate'].plot()
```

Plot success rate for each try as time increases.

```python
for k,g in data.groupby('Try'):
    plt.plot(g['Run'],g['SuccessRate'])
```
